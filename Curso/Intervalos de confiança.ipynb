```r
# Dados
set.seed(1234)
x <- rnorm(100)
```

```r
# Intevalo de confianÃ§a da mÃ©dia (Normal)
norm.interval = function(data, variance = var(data), conf.level = 0.95) { 
    z = qnorm((1 - conf.level)/2, lower.tail = FALSE)
    xbar = mean(data)
    sdx = sqrt(variance/length(data))
    c(xbar - z * sdx, xbar + z * sdx)
} 
```

```r
norm.interval(x)
```

```r
# Intevalo de confianÃ§a da variÃ¢ncia (Normal)
var.interval = function(data, conf.level = 0.95) { 
    df = length(data) - 1
    chilower = qchisq((1 - conf.level)/2, df) 
    chiupper = qchisq((1 - conf.level)/2, df, lower.tail = FALSE) 
    v = var(data) 
    c(df * v/chiupper, df * v/chilower) 
} 
```

```r
var.interval(x)
```

```r
# DistribuiÃ§Ã£o t
t.test(x,conf.level=0.95)
```

# Intervalos de confianÃ§a amostrais

```r
n.draw = 100 
mu = mean(x) 
n = length(x)
SD = sd(x) 
SD
```

```r
draws = matrix(rnorm(n.draw * n, mu, SD), n)
```

```r
# Quantos intervalos de confianÃ§a contÃ©m a mÃ©dia populacional
get.conf.int = function(x) t.test(x)$conf.int 
conf.int = apply(draws, 2, get.conf.int) 
sum(conf.int[1, ] <= mu & conf.int[2, ] >= mu)
```

```r
plot(range(conf.int), c(0, 1 + n.draw), type = "n", xlab = "Desvio da mÃ©dia", 
     ylab = "Amostragem") 
for (i in 1:n.draw) lines(conf.int[, i], rep(i, 2), lwd = 2) 
abline(v = 9, lwd = 2, lty = 2)
```

# Bootstrap

```r
install.packages("boot",dep=TRUE)
library(boot)
```

```r
# FunÃ§Ã£o onde O primeiro argumento Ã© o DataSet, e o segundo Ã­ndice das observaÃ§Ãµes neste DataSet.  
fc <- function(d, i){
    d2 <- d[i,]
    return(cor(d2$write, d2$math))
}
```

```r
hsb2<-read.table("https://stats.idre.ucla.edu/stat/data/hsb2.csv", sep=",", header=T)
hsb2
```

```r
bootcorr <- boot(hsb2, fc, R=500)
bootcorr
```

```r
summary(bootcorr)
```

```r
plot(bootcorr)
```

```r
boot.ci(boot.out = bootcorr, type = c("norm", "basic", "perc", "bca"))
```
