# Dados

```r
dados <- data.frame(VD=c(26,16,19,16,13,    18,21,18,11,21,
                         20,12,16,25,13,    15,15,22,14,17,    
                         10,24,17,17,14),
                    F1=factor(c("A","B","C","D","E",  "B","C","D","E","A",
                                "C","D","E","A","B",  "D","E","A","B","C",
                                "E","A","B","C","D")),
                    B1=factor(c("a","b","c","d","e",  "c","d","e","a","b",
                                "e","a","b","c","d",  "b","c","d","e","a",
                                "d","e","a","b","c")),
                    B2=gl(5,5), B3=gl(5,1,25))
```

```r
head(dados)
```

```r
sapply(dados,class)
```

## Modelo sem blocos

```r
m1 <- lm(VD ~ F1, dados)
anova(m1)
```

## Blocos ao acaso (paramÃ©trico)

```r
m1_ba <- lm(VD ~ F1 + B1, dados)
anova(m1_ba)
```
